```python

def sparse(file_name):
    index = 1
    x =[set() for i in range(5)]
    with open(file_name, 'r') as f:
        for row in f.readlines():
            if row != None:
                x[int(row)].add(index)
                index += 1 
    return x
def construct_graph(graph_name):
    y={}
    nodes = 0
    with open(graph_name, 'r') as f:
        row = f.readline()
        index = 1
        nodes = int(row.split()[0])
        for row in f.readlines():
            row = map(int,row.split())
            x=[]
            for item in row:
                x.append(item)
            y[index]=x
            index+=1
    return nodes,y
path_files = ['fast','eco','strong','highest']
data_files = ["kitti_07", "sphere2500","smallGrid3D","kitti_02","cubicle","input_INTEL_g2o",\
        "torus3D","kitti_09","city10000","parking-garage","kitti_06","kitti_05","ais2klinik","kitti_00","rim","CSAIL","kitti_08",\
        "grid3D","sphere_bignoise_vertex3","input_MITb_g2o","input_M3500_g2o"
        ]

for data_file in data_files:
    file_name  = f"./origin/{data_file}.graph"
    print(data_file)
    nodes,y = construct_graph(file_name)
    block_num = nodes // 5 +1
    cut_edges = [0 for i in range(5)]
    for i in range(nodes):
        for neighbour in y[i+1]:
            if (neighbour-1)//block_num != i//block_num:
                cut_edges[i//block_num]+=1
                cut_edges[(neighbour-1)//block_num]+=1

    # print(cut_edges)
    print(sum(cut_edges))   


    for path_file in path_files:
        cut_edges = [0 for i in range(5)]
        file_name = f"./{path_file}/{data_file}"
        x = sparse(file_name)
        for key,value in y.items():
            index = -1
            for item in range(len(x)):
                if key in x[item]:
                    index = item
                    break
            if index == -1:
                continue
            for neighbour in  value:
                if neighbour not in x[index]:
                    cut_edges[index] += 1
        
        # print(cut_edges)
        print(sum(cut_edges))



```
